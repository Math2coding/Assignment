     All the features have values as object type. So, we need to encode them inorder to feed into our ML models

```python
#One Hot Encoding
en_data=pd.get_dummies(data.iloc[:,1:24])
en_data
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
label_en=LabelEncoder()
y=label_en.fit_transform(data['class'])
```

```python
#Split data into train and test
x=en_data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,stratify=y,random_state=42)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
pred1=lr.predict(x_test)
from sklearn.model_selection import cross_val_score
score1= cross_val_score(lr, x_train, y_train, cv=10)
score1
```

```python
#Evaluation metric
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
accuracy1=accuracy_score(y_test,pred1)
precision1=precision_score(y_test,pred1)
recall1=recall_score(y_test,pred1)
fscore1=f1_score(y_test,pred1)
```

```python
#K-NN
from sklearn.neighbors import KNeighborsClassifier
acc_values=[]
neighbors=np.arange(3,15)
for k in neighbors:
    classifier=KNeighborsClassifier(n_neighbors=k,metric='minkowski')
    classifier.fit(x_train,y_train)
    pred2=classifier.predict(x_test)
    acc=accuracy_score(y_test,pred2)
    acc_values.append(acc)
plt.plot(neighbors,acc_values,'*-')
plt.xlabel('kvalue')
plt.ylabel('accuracy')
```

<u><b>Conclusion:</b></u>
         
         From the graph, we see that the highest accuracy is for k values k= 3,4,5,6,7,8,9.
         So taking one at random k=6

```python
classifier=KNeighborsClassifier(n_neighbors=6,metric='minkowski')
classifier.fit(x_train,y_train)
pred2=classifier.predict(x_test)
accuracy2=accuracy_score(y_test,pred2)
precision2=precision_score(y_test,pred2)
recall2=recall_score(y_test,pred2)
fscore2=f1_score(y_test,pred2)
from sklearn.model_selection import cross_val_score
score2= cross_val_score(classifier, x_train, y_train, cv=10)
score2
```

```python
#SVM
#SVM Linear
from sklearn.svm import SVC
svm_lr=SVC(kernel='linear')
svm_lr.fit(x_train,y_train)
pred3=svm_lr.predict(x_test)
accuracy3=accuracy_score(y_test,pred3)
precision3=precision_score(y_test,pred3)
recall3=recall_score(y_test,pred3)
fscore3=f1_score(y_test,pred3)
from sklearn.model_selection import cross_val_score
score3= cross_val_score(svm_lr, x_train, y_train, cv=10)
score3
```

```python
#SVM Poly
from sklearn.svm import SVC
svm_py=SVC(kernel='poly')
svm_py.fit(x_train,y_train)
pred4=svm_py.predict(x_test)
accuracy4=accuracy_score(y_test,pred4)
precision4=precision_score(y_test,pred4)
recall4=recall_score(y_test,pred4)
fscore4=f1_score(y_test,pred4)
from sklearn.model_selection import cross_val_score
score4= cross_val_score(svm_py, x_train, y_train, cv=10)
score4
```

```python
#SVM Radial
from sklearn.svm import SVC
svm_rbf=SVC(kernel='rbf')
svm_rbf.fit(x_train,y_train)
pred5=svm_rbf.predict(x_test)
accuracy5=accuracy_score(y_test,pred5)
precision5=precision_score(y_test,pred5)
recall5=recall_score(y_test,pred5)
fscore5=f1_score(y_test,pred5)
from sklearn.model_selection import cross_val_score
score5= cross_val_score(svm_rbf, x_train, y_train, cv=10)
score5
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
pred6=rf.predict(x_test)
accuracy6=accuracy_score(y_test,pred6)
precision6=precision_score(y_test,pred6)
recall6=recall_score(y_test,pred6)
fscore6=f1_score(y_test,pred6)
from sklearn.model_selection import cross_val_score
score6= cross_val_score(rf, x_train, y_train, cv=10)
score6
```

```python
#Naive Bayes
#Gaussian NB
from sklearn.naive_bayes import GaussianNB
classifier=GaussianNB()
classifier.fit(x_train,y_train)
pred7=classifier.predict(x_test)
accuracy7=accuracy_score(y_test,pred7)
precision7=precision_score(y_test,pred7)
recall7=recall_score(y_test,pred7)
fscore7=f1_score(y_test,pred7)
from sklearn.model_selection import cross_val_score
score7= cross_val_score(classifier, x_train, y_train, cv=10)
score7
```

```python
#Bernoulli NB
from sklearn.naive_bayes import BernoulliNB
classifier=BernoulliNB()
classifier.fit(x_train,y_train)
pred8=classifier.predict(x_test)
accuracy8=accuracy_score(y_test,pred8)
precision8=precision_score(y_test,pred8)
recall8=recall_score(y_test,pred8)
fscore8=f1_score(y_test,pred8)
from sklearn.model_selection import cross_val_score
score8= cross_val_score(classifier, x_train, y_train, cv=10)
score8
```

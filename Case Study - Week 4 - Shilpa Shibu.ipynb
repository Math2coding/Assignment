### Conclusion = No null values present

### Female and Male count

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.DataFrame(data)
Count=df['Buyer Gender'].value_counts()
Count
```

```python
Female=df['Buyer Gender']=='Female'
no_of_females=Female.sum()
no_of_females
```

```python
Male=df['Buyer Gender']=='Male'
no_of_males=Male.sum()
no_of_males
```

# Bar graph

```python
plt.bar(['Female','Male'],height=[no_of_females,no_of_males],color=['blue','orange'],width=0.7)
plt.grid()
plt.title('Female v/s Male Buyers participated in sales')
plt.ylabel('No.of participants')
```

# Top 5 cars according to their SP

### Top 5 data based on large S.P (Sale Price)

```python
large_SP=df.nlargest(5,'Sale Price')
large_SP
```

### Model of top 5 cars based on large S.P 

```python
Top_Cars=large_SP['Model']
Top_Cars
```

# Least 5 cars based on re-sell price

### Least 5 data based on resell price

```python
least_cars_RP=df.nsmallest(5,'Resell Price')
least_cars_RP
```

### Model of least 5 cars based on resell price  

```python
Least_Cars=least_cars_RP['Model']
Least_Cars
```
